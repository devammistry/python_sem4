```python
import matplotlib.pyplot as plt
import networkx as nx
G=nx.DiGraph()
G.add_edges_from([('a','b'),('b','d'),('d','a'),('c','d'), ('b', 'c')])
nx.draw_networkx(G, node_size=300, node_color='red', edge_color='blue', width=2)
plt.figsize=[200,200]
plt.show()
```

```python
# the list for person1 has the value on index 2 as 1 which means person1 followsperson3 and a directed edge should be
# added from person1 to person3
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
count=0
anslist=[]
for i in instagram.values():
    print(i)
    count+=1
    count1=0
    for j in i:
        count1+=1
        if j==1:
            anstuple=(("person"+str(count)), ("person"+str(count1)))
            anslist.append(anstuple)
G.add_edges_from(anslist)
nx.draw_networkx(G, node_size=3000, node_color='red', edge_color='blue', width=2)
plt.figsize=[400,400]
plt.show()
```
