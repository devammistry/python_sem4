```python
# box plot (whisker's plot)
import numpy as np
import pandas as pd
df=pd.read_csv('auto-mpg.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['mpg'], notch=True, widths=0.75)
plt.boxplot(df['acceleration'])
# notch shows CI: confidence interval
# narrow notch: high CI
# wide notch: more variability
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df[['mpg', 'acceleration']], notch=True, widths=0.75)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df[['mpg', 'acceleration', 'weight']], notch=True, widths=0.75)
```

```python
time=np.arange(12)   # 0 to 11 (both included)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time, income, color='green')
plt.plot(time, expense, color='red')
plt.show()
```

```python
time=np.arange(12)   # 0 to 11 (both included)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time, income, color='green')
plt.plot(time, expense, color='red')
plt.fill_between(time, income, expense, where=(income>expense), color='green')
plt.fill_between(time, income, expense, where=(income<expense), color='red')
plt.show()
```

```python
help(plt.fill_between)
```

```python
time=np.arange(12)   # 0 to 11 (both included)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time, income, color='green')
plt.plot(time, expense, color='red')
plt.fill_between(time, income, expense, where=(income>expense), color='green', interpolate=True, alpha=0.3, label='profit')
plt.fill_between(time, income, expense, where=(income<expense), color='red', interpolate=True, alpha=0.3, label='loss')
plt.legend()
plt.show()
```

```python
# stack area chart
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
plt.stackplot(x,y1,y2,y3, labels=['a', 'b', 'c'], colors=['red', 'blue', 'green'], alpha=0.3)
plt.legend(loc='upper left')
plt.show()
```

```python
# regression plots
import seaborn as sns
df=pd.read_csv('auto-mpg.csv')
df.corr()
```

```python
sns.regplot(x='mpg', y='displacement', data=df)
```

```python
sns.regplot(x='cylinders', y='displacement', data=df)
```

```python
help(sns.regplot)
```

```python
sns.regplot(x='cylinders', y='displacement', data=df, color='red', marker='*', scatter_kws={'s':200})
plt.xlabel('cylinders')
plt.ylabel('disp')
plt.title('cyl vs disp')
```
