```python
import pandas as pd
```

```python
df=pd.read_csv("temp.csv")
df
```

```python
df=df.drop("sno", axis=1)
df
```

```python
y=df["Pressure"]
x=df.drop("Pressure", axis=1)
print(x.shape, y.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
x_poly
```

```python
print(x_poly.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_poly, y, test_size=0.2)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
```

```python
df1=pd.read_csv("Real estate.csv")
df1
```

```python
df1=df1.drop("No", axis=1)
df1
```

```python
y=df1["Y house price of unit area"]
x=df1.drop("Y house price of unit area", axis=1)
print(x.shape, y.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
x_poly
```

```python
print(x_poly.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_poly, y, test_size=0.2, random_state=2)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
```
