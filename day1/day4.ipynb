```python
# IQR: inter quantile range
```

```python
def find_outliers(ds, col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    iqr=q3-q1
    low_val=q1-1.5*iqr
    print(low_val)
    high_val=q3+1.5*iqr
    print(high_val)
    ds=ds.loc[(ds[col]<low_val)|(ds[col]>high_val)]
    return ds
```

```python
find_outliers(df, 'mpg')
```
