```python
sales_data=pd.DataFrame({'name':['a',np.nan, np.nan, 'd', 'e'], 
                         'region':['ma', np.nan, 'mp', 'gu', np.nan], 
                         'sales':[10, np.nan, 30, np.nan, 50], 
                         'expense':[50, np.nan, 70, np.nan, 90]})
sales_data
```

```python
sales_data.isna()         #nan che k nai
```

```python
sales_data.isna().sum()       # count of nan in columns
```

```python
sales_data.dropna()          # ekpan na vagar ni row
```

```python
sales_data.dropna(how='all')         # badha nan hashe a kadhi dese
```

```python
sales_data.dropna(how='any')         # ekpan na vagar ni row     same as dropna()
```

```python
sales_data.dropna(subset=['sales'])
```

```python
sales_data.dropna(subset=['sales', 'region'])
```
