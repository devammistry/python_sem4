```python
import pandas as pd
```

```python
data = {
    'calories': [420, 380, 390],
    'duration': [50, 40, 45]
}
df=pd.DataFrame(data, index=["first-day", "second-day", "third-day"])
print(df)
print(df['calories'][0])
print(df['calories']['first-day'])
# print(df['calories'].loc[0])      error
# loc (location) accepts labels (rowname) as well as int
```

```python
data = {
    'calories': [420, 380, 390],
    'duration': [50, 40, 45]
}
df=pd.DataFrame(data)
print(df)
print(df['calories'].iloc[0])                # int loc
```

```python
data = {
    'calories': [420, 380, 390],
    'duration': [50, 40, 45]
}
df=pd.DataFrame(data, index=["first-day", "second-day", "third-day"])
print(df)
print(df['calories'].iloc[0])                # int loc
print(df['calories'].iloc["first-day"])      # iloc accepts only int
```
