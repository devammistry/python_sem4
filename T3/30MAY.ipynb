```python
# http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={API key}
api_key='cd361448c88fdaf71fcbfd2e2bcde868'
lat, lon=get_coordinates("london")
url='https://api.openweathermap.org/data/2.5/air_pollution?lat='+str(lat)+'&lon='+str(lon)+'&appid='+api_key
response=requests.get(url).json()
aqi=response["list"][0]['main']["aqi"]
d={1:"Good",2:"Fair",3:"Moderate",4:"Poor",5:"Very Poor"}
print(d[aqi])
# https://openweathermap.org/api/air-pollution
```

```python
# api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API key}
api_key='cd361448c88fdaf71fcbfd2e2bcde868'
lat, lon=get_coordinates("ahmedabad")
url='https://api.openweathermap.org/data/2.5/forecast?lat='+str(lat)+'&lon='+str(lon)+'&appid='+api_key
response=requests.get(url).json()
# print(json.dumps(response,indent=5))
d={"date_time":[], "temp":[], "pressure":[], "humidity":[], "description":[]}
for i in response["list"]:
  d["date_time"].append(i["dt_txt"])
  d["temp"].append(i["main"]["temp"])
  d["pressure"].append(i["main"]["pressure"])
  d["humidity"].append(i["main"]["humidity"])
  d["description"].append(i["weather"][0]["description"])
import pandas as pd
df=pd.DataFrame(d)
df
# https://openweathermap.org/forecast5
# df.to_csv("weather.csv",index=False)
```
