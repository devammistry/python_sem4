```python
import requests
import bs4
f=open('Today 52 Week Low BSE_NSE Stocks Companies List â€“ Ticker.html', encoding="utf-8")
soup=bs4.BeautifulSoup(f, 'html.parser')
print(soup.prettify())
```

```python
dic={'S.No.':[],'Company':[], 'price Rs:':[], 'Day low Rs.':[]}
```

```python
transcript=soup.find_all('td', class_='left')
print(transcript)
```

```python
for i in transcript:
  b=i.text.strip()
  dic['Company'].append(b)
```

```python
transcript=soup.find_all('td', class_='Number')
print(transcript)
```

```python
for i in transcript:
  b=i.text.strip()
  dic['price Rs:'].append(b)
```

```python
transcript=soup.find_all('span', class_='Number')
print(transcript)
```

```python
for i in transcript:
  b=i.text.strip()
  dic['Day low Rs.'].append(b)
```

```python
sr=soup.find_all('tr')
for i in sr[1:]:
  b=i.text[0:3].strip()
  dic['S.No.'].append(b)
```

```python
print(dic)
```

```python
import pandas as pd
df=pd.DataFrame(dic)
```

```python
f=open('Certified used Mercedes-Benz for sale.html', encoding="utf-8")
soup=bs4.BeautifulSoup(f, 'html.parser')
print(soup.prettify())
```

```python
dic={'Name':[],'Mileage':[], 'Dealer Name':[], 'Review Count':[], 'Price':[]}
```

```python
for i in transcript[1:]:
  b=i.text.strip()
  dic['Mileage'].append(b)
```

```python
print(dic)
```

```python
transcript=soup.find_all('div', class_='dealer-name')
print(transcript)
```

```python
for i in transcript:
  b=i.text.strip()
  dic['Dealer Name'].append(b)
```

```python
transcript=soup.find_all('span', class_='sds-rating__link sds-button-link')
print(transcript)
```

```python
for i in transcript:
  b=i.text[1:4].strip()
  dic['Review Count'].append(b)
```

```python
transcript=soup.find_all('span', class_='primary-price')
print(transcript)
```
